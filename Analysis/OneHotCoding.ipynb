```python
import pandas as pd
```

```python
data = pd.read_csv('DATA/act_labeling_data_fix.csv',encoding='cp949')
```

```python
data['BIZNML_x'].value_counts()
```

```python
kind_job = []
for a in range(0,len(data)):
    kind_job.append(data['BIZNML_x'][a][0])
```

```python
data['KIND_JOB']=kind_job
```

```python
data['KIND_JOB'].value_counts()
```

```python
b= pd.get_dummies(data['KIND_JOB'],prefix='KIND_JOB')
```

```python
aaa = pd.concat([data,b],axis=1)
```
