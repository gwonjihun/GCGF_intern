```python
df = pd.read_csv("RISKCHECK\onehot.csv",encoding="cp949")
```

```python
df.columns
```

```python
df['YEAR']=0
df['MONTH']=0
df["DAY"]=0
```

```python
year=[]
month=[]
day=[]
for a in range(0,len(df)):
    year.append(df["보증일자"][a]//10000)
    month.append((df["보증일자"][a]%10000)//100)
    day.append(df["보증일자"][a]%100)
```

```python
df['YEAR']=year
df['MONTH']=month
df["DAY"]=day
```

```python
df['quarter']=0
```

```python
for a in range(0,len(df)):
    if df['MONTH'][a] <=3:
        df['quarter'][a]= 1
    elif df['MONTH'][a] <=6:
        df['quarter'][a]= 2
    elif df['MONTH'][a] <=9:
        df['quarter'][a]= 3
    else:
        df['quarter'][a]= 4
```

```python
day_df = pd.read_csv('RISKCHECK/Resource/일별_코스피_환율.csv',encoding='cp949')
```

```python
day_df
```

```python
year=[]
month=[]
day=[]
```

```python
for a in range(0,len(day_df)):
    year.append(int(day_df['년'][a][0:4]))
    month.append(int(day_df['년'][a][5:7]))
    day.append(int(day_df['년'][a][-2:]))
```

```python
day_df['YEAR']= year
day_df['MONTH']= month
day_df['DAY'] = day
```

```python
day_df
```

```python
day_df.to_csv('RISKCHECK/Resource/일별_코스피_환율_분류.csv',encoding='cp949',index=False)
```
